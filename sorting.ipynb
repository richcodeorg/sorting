## What is a Lambda Function?
A lambda function is a small, anonymous (unnamed) function in Python defined using the lambda keyword.

- It can take any number of arguments but can only have one expression.

- It is used for short, simple operations where a full function definition is unnecessary.

## Bubble Sort

Bubble Sort is a simple sorting algorithm that **repeatedly steps through a list**, compares **adjacent elements**, and **swaps them if they are in the wrong order**. This process repeats until the entire list is sorted.

It‚Äôs called "Bubble Sort" because the largest elements "bubble up" to the end of the list, just like bubbles rising in water!

## How Bubble Sort Works
1. Start at the beginning of the list.

2. Compare the first two elements.

    If the first is larger, swap them.

    Otherwise, move to the next pair.

3. Repeat for every pair until the end.

4. After each pass, the largest unsorted element "bubbles up" to its correct position.

5. Repeat the process until no more swaps are needed.


# **Bubble Sort Example: Sorting `[5, 3, 8, 4, 2]`**

## How Bubble Sort Works Step-by-Step

| **Pass** | **Comparison & Action**               | **List State**       |
|----------|---------------------------------------|----------------------|
| **1st**  | Compare (5,3) ‚Üí Swap                  | `[3, 5, 8, 4, 2]`    |
|          | Compare (5,8) ‚Üí No swap               | `[3, 5, 8, 4, 2]`    |
|          | Compare (8,4) ‚Üí Swap                  | `[3, 5, 4, 8, 2]`    |
|          | Compare (8,2) ‚Üí Swap                  | `[3, 5, 4, 2, 8]`    |
| **2nd**  | Compare (3,5) ‚Üí No swap               | `[3, 5, 4, 2, 8]`    |
|          | Compare (5,4) ‚Üí Swap                  | `[3, 4, 5, 2, 8]`    |
|          | Compare (5,2) ‚Üí Swap                  | `[3, 4, 2, 5, 8]`    |
| **3rd**  | Compare (3,4) ‚Üí No swap               | `[3, 4, 2, 5, 8]`    |
|          | Compare (4,2) ‚Üí Swap                  | `[3, 2, 4, 5, 8]`    |
| **4th**  | Compare (3,2) ‚Üí Swap                  | `[2, 3, 4, 5, 8]`    |
| **Done!**| No more swaps needed! ‚úÖ               | `[2, 3, 4, 5, 8]`    |


```python
import time
from random import randint

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def bubbleSort(array):
    import time
    start = time.time()
    for i in range(len(array)): # n pass
        is_sorted = True  # initialize is_sorted
        for j in range(1, len(array) - i):
            if (array[j] < array[j - 1]):
                # swap
                array[j], array[j - 1] = array[j - 1], array[j]
                is_sorted = False

        if (is_sorted): break
    t = time.time() - start
    return len(array), t

l = [1, 3, 5, 7, 9, 2, 4, 6, 8, 0]
bubbleSort(l)
print(l)
```

### Explanation of is_sorted in Optimized Bubble Sort

The is_sorted flag is an important optimization in the bubble sort algorithm. Let me explain why it's needed and how it works:

### Why We Need is_sorted
1. Early Termination:

  Without is_sorted, the algorithm always makes n passes through the array (where n is the array length)

  With is_sorted, we can stop early if the array becomes sorted before all passes complete

2. Best-Case Improvement:

For an already-sorted array, standard bubble sort still does O(n¬≤) comparisons

With is_sorted, it does just O(n) comparisons (one pass) and terminates.

### Example with Your Input [1, 3, 5, 7, 9, 2, 4, 6, 8, 0]
1. First pass:

    Makes several swaps (9‚Üî2, 9‚Üî4, etc.)

    is_sorted becomes False due to swaps

2. Subsequent passes:

    Continue making swaps

3. Final pass when sorted:

    No swaps occur ‚Üí is_sorted remains True

    Loop breaks early instead of continuing unnecessary passes

### This optimization is particularly valuable for:

- Nearly-sorted data

- Small datasets where early termination is likely

- Cases where you might re-sort already-sorted data

The **is_sorted** flag makes bubble sort more practical for real-world use cases where data often isn't completely random.

## Understanding Big O Notation (Time Complexity)

Big O notation describes how an algorithm's runtime or memory usage grows relative to input size (n). It answers:

"If the input grows 10x, how much slower will my code run?"

**1. What Does Big O Measure?**

- Worst-case performance (upper bound)

- Ignores constants (e.g., O(2n) ‚Üí O(n))

- Focuses on dominant terms (e.g., O(n¬≤ + n) ‚Üí O(n¬≤))

# Common Time Complexities

## üèÜ Fastest to Slowest

| Complexity   | Name          | Example                          | When n=1,000       |
|--------------|---------------|----------------------------------|--------------------|
| **O(1)**     | Constant      | Array lookup (`arr[0]`)          | 1 operation        |
| **O(log n)** | Logarithmic   | Binary search                    | ~10 operations     |
| **O(n)**     | Linear        | Looping through an array         | 1,000 operations   |
| **O(n log n)** | Linearithmic | Merge Sort, Quick Sort           | ~10,000 operations |
| **O(n¬≤)**    | Quadratic     | Nested loops (Bubble Sort)       | 1,000,000 ops     |
| **O(2‚Åø)**    | Exponential   | Brute-force password cracking    | 1e+12 operations  |
| **O(n!)**    | Factorial     | Traveling Salesman (naive)       | 4e+2567 ops*      |

\* *For n=1000, n! has 2568 digits - practically uncomputable*


## Big-O Cheat Sheet
- O(1)     = Constant
- O(log n) = Logarithmic
- O(n)     = Linear
- O(n¬≤)    = Quadratic
- O(2‚Åø)    = Exponential
- O(n!)    = Factorial

## Selection Sort

Selection Sort is a simple comparison-based sorting algorithm that:

- Divides the array into a sorted part (left) and unsorted part (right)

- Repeatedly selects the smallest element from the unsorted part and swaps it with the leftmost unsorted element

- Has O(n¬≤) time complexity in all cases (worst, average, best)

- Uses O(1) extra space (in-place sorting)

## **Selection Sort Step-by-Step Example**  
### Sorting `[29, 10, 14, 37, 13]`

| Pass      | Action                          | Array State               |
|-----------|---------------------------------|---------------------------|
| **Initial** | Unsorted array                 | `[29, 10, 14, 37, 13]`    |
| **Pass 1** | Find min `10` ‚Üí swap with `29` | `[`**10**`, 29, 14, 37, 13]` |
| **Pass 2** | Find min `13` ‚Üí swap with `29` | `[10,` **13**`, 14, 37, 29]` |
| **Pass 3** | Find min `14` ‚Üí already in place | `[10, 13,` **14**`, 37, 29]` |
| **Pass 4** | Find min `29` ‚Üí swap with `37` | `[10, 13, 14,` **29**`, 37]` |
| **Done!**  | Array is sorted                | `[10, 13, 14, 29, 37]` ‚úÖ |

**Key Observations:**
- Each pass selects the smallest remaining element (highlighted in **bold**)
- Swaps place one element in its correct position per iteration
- After **n-1 passes**, the array is fully sorted

```python
def selection_sort(items):
    start = time.time()
    for i in range(len(items)):   # n
        min_index = i   #index
        for j in range(i + 1, len(items)):  # n
            if (items[j] < items[min_index]):
                min_index = j

        items[i], items[min_index] = items[min_index], items[i]
    t = time.time() - start
    return len(items), t
```

### Time Complexity Analysis**

| Case          | Time Complexity | Explanation                          |
|---------------|-----------------|--------------------------------------|
| **Worst Case** | O(n¬≤)          | Array is in reverse order            |
| **Average Case** | O(n¬≤)        | Randomly ordered elements            |
| **Best Case** | O(n¬≤)          | Already sorted (still checks all elements) |

### **Why O(n¬≤)?**
- **Outer loop** runs **n times** (once for each element)
- **Inner loop** runs **n-i times** (finds minimum in unsorted portion)
- Total comparisons: n + (n-1) + (n-2) + ... + 1 = n(n-1)/2 ‚Üí **O(n¬≤)**
- Same complexity for all cases because it always performs full scans

### Selection Sort vs. Bubble Sort Comparison**

| Feature          | Selection Sort               | Bubble Sort                  |
|------------------|------------------------------|------------------------------|
| **Time Complexity** | O(n¬≤) (all cases)          | O(n¬≤) worst/average, O(n) best (optimized) |
| **Swaps**        | O(n) (exactly n-1 swaps)     | O(n¬≤) (up to n¬≤ swaps)       |
| **Adaptivity**   | ‚ùå Not adaptive              | ‚úÖ Adaptive (with `is_sorted` flag) |
| **Stability**    | ‚ùå Not stable                | ‚úÖ Stable                     |
| **Memory**       | O(1) (in-place)              | O(1) (in-place)              |
| **Best For**     | Small datasets with expensive writes | Nearly-sorted data       |

### **Key Differences Explained:**
1. **Swap Efficiency**:
   - Selection Sort makes *at most* **n-1 swaps** (good when writes are expensive)
   - Bubble Sort can require **n¬≤ swaps** in worst case

2. **Adaptivity**:
   - Bubble Sort can **early-terminate** if array is sorted (adaptive)
   - Selection Sort **always scans entire array**

3. **Stability**:
   - Bubble Sort preserves order of equal elements (stable)
   - Selection Sort may change order (unstable due to swaps)

### **When to Use Which?**
- Use **Selection Sort** when:
  - You have limited write operations (e.g., flash memory)
  - Dataset is small and simplicity is preferred

- Use **Bubble Sort** when:
  - Data is nearly sorted
  - Stability is required
  - Implementations can use early termination

## Insertion Sort

Insertion Sort is a simple comparison-based sorting algorithm that:

- Builds the final sorted array one element at a time

- Works similarly to how you might sort playing cards in your hands

- Has O(n¬≤) worst-case time complexity, but O(n) best-case for nearly sorted data

- Is stable (maintains relative order of equal elements)

- Uses O(1) extra space (in-place sorting)

## **How Insertion Sort Works**

### **Step-by-Step Example (Sorting `[5, 2, 4, 6, 1, 3]`)**

| **Pass** | **Action**                          | **Array State**          |
|----------|-------------------------------------|--------------------------|
| Initial  | Unsorted array                      | `[5, 2, 4, 6, 1, 3]`     |
| Pass 1   | Insert 2 before 5                   | `[2, 5, 4, 6, 1, 3]`     |
| Pass 2   | Insert 4 between 2 and 5            | `[2, 4, 5, 6, 1, 3]`     |
| Pass 3   | 6 is already in correct position    | `[2, 4, 5, 6, 1, 3]`     |
| Pass 4   | Insert 1 at the beginning           | `[1, 2, 4, 5, 6, 3]`     |
| Pass 5   | Insert 3 between 2 and 4            | `[1, 2, 3, 4, 5, 6]`     |
| Done!    | Array is sorted                     | `[1, 2, 3, 4, 5, 6]` ‚úÖ  |

**Visualization**:

- [5] ‚Üê Start with first element sorted
- [2, 5] ‚Üê Insert 2 before 5
- [2, 4, 5] ‚Üê Insert 4 between 2 and 5
- [2, 4, 5, 6] ‚Üê 6 stays
- [1, 2, 4, 5, 6] ‚Üê Insert 1 at beginning
- [1, 2, 3, 4, 5, 6] ‚Üê Final sorted array

## How Insertion Sort Works**

![Insertion Sort Animation](https://upload.wikimedia.org/wikipedia/commons/0/0f/Insertion-sort-example-300px.gif)

*Visualization: Elements are taken one by one and inserted into their correct position*




```python
import time

def insert_sort(items):
    start = time.time()
    for i in range(1, len(items)):
        current_index = i
        while current_index > 0 and items[current_index - 1] > items[current_index]:
            items[current_index - 1], items[current_index] = items[current_index], items[current_index - 1]
            current_index -= 1
    t = time.time() - start
    return len(items), t
```

| **Case**      | Time Complexity | Description                          |
|---------------|-----------------|--------------------------------------|
| **Worst Case** | O(n¬≤)          | Array is in reverse order            |
| **Average Case** | O(n¬≤)        | Randomly ordered elements            |

**Explanation**:

- For already sorted data: only 1 comparison per element ‚Üí O(n)
- Average case requires ‚âà n¬≤/4 comparisons

**Mathematical Proof**:
- Worst case comparisons: 1 + 2 + ... + (n-1) = n(n-1)/2 ‚Üí O(n¬≤)
- Best case comparisons: (n-1) ‚Üí O(n)

| **Space**         | O(1)               | O(1)               | O(1)              | O(log n)       |
| **Stable**        | ‚úÖ Yes             | ‚ùå No              | ‚úÖ Yes            | ‚ùå Usually no  |
| **Adaptive**      | ‚úÖ Yes             | ‚ùå No              | ‚úÖ Yes            | ‚ùå No          |
| **Good For**      | Small/nearly-sorted| Small datasets     | Educational       | General purpose|

**Key Advantages**:
- Insertion Sort outperforms others for:
  - Very small datasets (n < 10)
  - Nearly-sorted data (O(n) time)
  - Online sorting (data arriving sequentially)

```python
# Generate test cases
random_lists = [generate_random_array(100 * n) for n in range(1, 40)]

# Time all algorithms
insert_results = [insert_sort(l.copy()) for l in random_lists]
select_results = [selection_sort(l.copy()) for l in random_lists]
bubble_results = [bubbleSort(l.copy()) for l in random_lists]

# Prepare data for plotting
x = [r[0] for r in insert_results]  # Sizes

plt.figure(figsize=(10, 6))
plt.plot(x, [r[1] for r in insert_results], label='Insertion Sort')
plt.plot(x, [r[1] for r in select_results], label='Selection Sort')
plt.plot(x, [r[1] for r in bubble_results], label='Bubble Sort')

plt.title('Sorting Algorithm Performance Comparison')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import copy


# Test with nearly-sorted data
def generate_nearly_sorted(size):
    arr = list(range(size))
    # Add slight disorder
    for _ in range(size//10):
        i, j = random.sample(range(size), 2)
        arr[i], arr[j] = arr[j], arr[i]
    return arr

# Generate test cases (nearly sorted)
nearly_sorted_lists = [generate_nearly_sorted(100 * n) for n in range(1, 40)]

# Time all algorithms (using deep copies)
def benchmark(algorithm, data):
    return [algorithm(copy.deepcopy(l)) for l in data]

insert_results = benchmark(insert_sort, nearly_sorted_lists)
select_results = benchmark(selection_sort, nearly_sorted_lists)
bubble_results = benchmark(bubbleSort, nearly_sorted_lists)

# Prepare data for plotting
x = [r[0] for r in insert_results]  # Sizes

# Create plot
plt.figure(figsize=(12, 7))
plt.plot(x, [r[1] for r in insert_results], 'g-', label='Insertion Sort')
plt.plot(x, [r[1] for r in select_results], 'b-', label='Selection Sort')
plt.plot(x, [r[1] for r in bubble_results], 'r-', label='Bubble Sort')

plt.title('Sorting Algorithm Performance on Nearly-Sorted Data', fontsize=14)
plt.xlabel('Array Size', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Add annotations
plt.annotate('Insertion Sort excels\non nearly-sorted data',
             xy=(1000, insert_results[9][1]),
             xytext=(1200, insert_results[9][1]*0.5),
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=10)

plt.show()
```
